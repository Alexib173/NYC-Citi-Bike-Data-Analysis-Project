# Phase 2: Data Cleaning
### In this section, I will prepare the data for analysis by:
### 1. Converting data types (timestamps).
### 2. Creating new columns to help with analysis (duration, day of week).
### 3. Handling missing values (NaNs).
### 4. Removing logical outliers (trips that are too short or too long).

```python
'''
Data Type Conversion
Convert `started_at` and `ended_at` to datetime objects.
Create a `duration_minutes` column.
The raw dataset has start/end times, but not the actual length of the trip. This is crucial for our analysis.
'''
# 1. Convert timestamp columns to datetime objects
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

# 2. Create a Duration Column (in Minutes)
df['duration_minutes'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60

# 3. Create 'Day of Week' and 'Hour' columns for temporal analysis
df['day_of_week'] = df['started_at'].dt.day_name()
df['start_hour'] = df['started_at'].dt.hour
```

```python
# Verify the new columns were created correctly
df[['started_at', 'duration_minutes', 'day_of_week']].head()
```

```python
# Verify the data types are now correct
df.dtypes
```

```python
# I will check which columns have missing data.

# Drop rows where start_station_name or end_station_name is missing
# We use 'inplace=True' to modify the dataframe permanently
df.dropna(subset=['start_station_name', 'end_station_name','end_lat', 'end_lng','end_station_id' ], inplace=True)

# after
df.isnull().sum()
```

```python
#  See statistics BEFORE removing outliers
# Look at the 'min' and 'max' values here. 
display(df['duration_minutes'].describe())
```

```python
# Count how many rows are actually outliers
# Trips < 1 minute or > 180 minutes (3 hours)
outliers = df[(df['duration_minutes'] < 1) | (df['duration_minutes'] > 180)]
print(f"Number of outliers: {len(outliers)}")
print(f"Percentage of dataset: {(len(outliers) / len(df)) * 100:.2f}%")
```

```python
# Define criteria for keeping data
# Keep trips greater than 1 minute AND less than 180 minutes (3 hours)
clean_df = df[(df['duration_minutes'] > 1) & (df['duration_minutes'] < 180)]

# Update the main dataframe
df = clean_df

# Check stats after outlier removal
df['duration_minutes'].describe()
```

## Phase 3: Data Analysis & Visualization
### Now that the data is clean, I will answer 4 questions using aggregation and visualization.

### Question 1: Peak Utilization Analysis 
### What are the peak usage hours during weekdays versus weekends?

```python
#1. Create a "Weekend" vs "Weekday" column
df['is_weekend'] = df['day_of_week'].isin(['Saturday', 'Sunday'])
df['day_type'] = df['is_weekend'].map({True: 'Weekend', False: 'Weekday'})

```

```python
# 2. Aggregate: Count trips by Hour and Day Type
hourly_counts = df.groupby(['start_hour', 'day_type']).size().reset_index(name='trip_count')
hourly_counts.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=hourly_counts, x='start_hour', y='trip_count', hue='day_type', marker='o')

plt.title('Peak Usage Hours: Weekdays vs. Weekends')
plt.xlabel('Hour of Day (0-23)')
plt.ylabel('Number of Trips')
plt.xticks(range(0, 24))
plt.show()
```

### As the results show, weekend hours are significantly greater than weekday hours.

### Question 2: User Segmentation
### How does the average trip duration differ between Members and Casual users?

```python
# 1. Aggregate: Calculate Average Duration by User Type
user_stats = df.groupby('member_casual')['duration_minutes'].mean().reset_index()

# Display the specific numbers
display(user_stats)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=user_stats, x='member_casual', y='duration_minutes', palette='viridis')

plt.title('Average Trip Duration: Members vs. Casual Users')
plt.xlabel('User Type')
plt.ylabel('Average Duration (Minutes)')
plt.show()
```

### This result surprised me, as I initially expected members to use the bikes more frequently than casual riders. This insight suggests that offering discounts or additional benefits to members could encourage longer ride times an idea that could be valuable to offer customers.

### Question 3: Station Imbalance Analysis
### Which stations have the highest "Start" vs. "End" disparity?

```python
# 1. Count starts per station
starts = df['start_station_name'].value_counts().reset_index()
starts.columns = ['station_name', 'start_count']

# 2. Count ends per station
ends = df['end_station_name'].value_counts().reset_index()
ends.columns = ['station_name', 'end_count']

# 3. Merge the two frames
station_flow = pd.merge(starts, ends, on='station_name', how='outer').fillna(0)

# 4. Calculate Net Flow (Ends - Starts)
station_flow['net_flow'] = station_flow['end_count'] - station_flow['start_count']
```

```python
# Get Top 5 
most_full = station_flow.nlargest(5, 'net_flow')
most_empty = station_flow.nsmallest(5, 'net_flow')

# Stations Accumulating Bikes (Need Trucks to Remove
display(most_full[['station_name', 'net_flow']])

#Stations Draining Bikes (Need Trucks to Refill
display(most_empty[['station_name', 'net_flow']])
```

```python
# Visualization
plt.figure(figsize=(12, 6))
# Combine them for one plot
imbalance = pd.concat([most_full, most_empty])
sns.barplot(data=imbalance, x='net_flow', y='station_name')
plt.title('Top 5 Most Imbalanced Stations (Net Flow)')
plt.xlabel('Net Flow (Arriving - Leaving)')
plt.ylabel('Station Name')
plt.axvline(0, color='black', linewidth=1)
plt.show()
```

### Positive number = Bikes piling up (Net Accumulation)
### Negative number = Bikes draining away (Net Loss)

### The left station are the ones that constanly run out of bikes while the rigth are where everyone rides to.

### Question 4: Route Popularity Analysis
### What are the most popular routes (Start to End)?

```python
# 1. Create a Route column
df['route'] = df['start_station_name'] + " to " + df['end_station_name']

# 2. Count the routes
top_routes = df['route'].value_counts().head(10).reset_index()
top_routes.columns = ['route', 'trip_count']

# Top 10 Most Popular Commuter Routes
display(top_routes)
```

```python
# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(data=top_routes, x='trip_count', y='route')
plt.title('Top 10 Most Popular Routes')
plt.xlabel('Number of Trips')
plt.ylabel('Routes Start -> End ')
plt.show()
```

### Business Value, Identifying these high-traffic corridors helps the city plan where to build protected bike lanes. It also helps Lyft know where to deploy users paid to move bikes to move inventory along these specific lines.
